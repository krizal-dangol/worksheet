```python
import numpy as np

# 1.
empty_array = np.empty((2, 2))
print("Empty Array (2x2):\n", empty_array)

# 2.
ones_array = np.ones((4, 2))
print("\nAll Ones Array (4x2):\n", ones_array)

# 3.
full_array = np.full((3, 3), 7)
print("\nFull Array (3x3 filled with 7):\n", full_array)

# 4.
given_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(given_array)
print("\nZeros Like Given Array:\n", zeros_like_array)

# 5.
ones_like_array = np.ones_like(given_array)
print("\nOnes Like Given Array:\n", ones_like_array)

# 6.
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nConverted NumPy Array:\n", numpy_array)
```

```python
import numpy as np
# 1.
array1 = np.arange(10, 50)
print("Array from 10 to 49:\n", array1)

# 2.
matrix3x3 = np.arange(9).reshape(3, 3)
print("\n3x3 Matrix (0 to 8):\n", matrix3x3)

# 3.
identity_matrix = np.eye(3)
print("\n3x3 Identity Matrix:\n", identity_matrix)

# 4.
random_array = np.random.random(30)
mean_value = random_array.mean()
print("\nRandom Array of size 30:\n", random_array)
print("Mean of array:", mean_value)

# 5.
random_10x10 = np.random.random((10, 10))
min_value = random_10x10.min()
max_value = random_10x10.max()
print("\n10x10 Random Array:\n", random_10x10)
print("Min value:", min_value, "Max value:", max_value)

# 6.
zero_array = np.zeros(10)
zero_array[4] = 1  # index starts from 0
print("\nZero array with 5th element as 1:\n", zero_array)

# 7.
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\nReversed array:\n", reversed_arr)

# 8.
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n2D array with 1 on border and 0 inside:\n", border_array)

# 9.
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n8x8 Checkerboard pattern:\n", checkerboard)
```

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1.
add_xy = x + y
print("x + y:\n", add_xy)

# 2.
sub_xy = x - y
print("\nx - y:\n", sub_xy)

# 3.
mult_x = x * 3
print("\nx * 3:\n", mult_x)

# 4.
square_x = x ** 2
print("\nx squared:\n", square_x)

# 5.
dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)

print("\nDot product v.w:", dot_vw)
print("Dot product x.v:\n", dot_xv)
print("Dot product x.y:\n", dot_xy)

# 6.
concat_row_xy = np.concatenate((x, y), axis=0)
concat_col_vw = np.concatenate((v.reshape(-1, 1), w.reshape(-1, 1)), axis=1)

print("\nConcatenate x and y along rows:\n", concat_row_xy)
print("Concatenate v and w along columns:\n", concat_col_vw)

# 7.
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("\nConcatenate x and v:\n", concat_xv)
except ValueError as e:
    print("\nError concatenating x and v:", e)
    print("Explanation: x is a 2x2 matrix and v is a 1D array of length 2. Dimensions do not match along axis=1 for row-wise concatenation, hence it throws an error.")
```

```python
import numpy as np

# Given Arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1.
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("1. A × A⁻¹:\n", identity_check)


# 2.
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("AB equal to BA?", np.array_equal(AB, BA))


# 3.
AB_transpose = (np.dot(A, B)).T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)ᵀ:\n", AB_transpose)
print("BᵀAᵀ:\n", BT_AT)
print("Are they equal?", np.array_equal(AB_transpose, BT_AT))


# 4.
# Equations:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Using inverse
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("\n4. Solution using inverse [x, y, z]:", X)

# Optional: Solve using np.linalg.solve (more efficient)
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve [x, y, z]:", X_solve)
```
